```python
import pandas as pd
```

```python
df = pd.read_html('http://digidb.io/digimon-list/')[0]
df
```

```python
df.loc[df['Type'] == 'Vaccine']
```

```python
df.loc[(df['Type'] == 'Vaccine') & (df['HP'] > 1100)]
```

```python
HP_Mean = df['HP'].mean()
```

```python
by_virus = df.loc[df['Type'] == 'Virus']
by_virus.head()
```

```python
hp_mean = by_virus['HP'].mean()
hp_mean
```

```python
sp_mean = by_virus['Spd'].mean()
sp_mean
```

```python
asd = df.loc[(df['Type'] == 'Virus') & 
       (df['HP'] > hp_mean) &
       (df['Memory'] > 18) & 
       (df['Atk'] > 100) &
       (df['Spd'] >= sp_mean)]
asd
```

```python
asd['Digimon'].count()
```

```python
asd.to_csv('Digimon.csv', index=False)
```
