```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r'C:\Users\dheof\Desktop\anime.csv')
df
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df = df.reset_index(drop=True)
```

```python
# Generate item feature matrix
vect = CountVectorizer(tokenizer=lambda x:x.split(', '))
df_genre = vect.fit_transform(df['genre'])
df_genre = pd.DataFrame(df_genre.toarray(), columns=vect.get_feature_names())
df_genre = pd.concat([df[['anime_id', 'name', 'rating']], df_genre], axis=1)
df_genre
```

```python
df_item_features = df_genre[df_genre['name'].isin([
    'Kimi no Na wa.', 
    'Fullmetal Alchemist: Brotherhood', 
    'Gintama°'])]
df_item_features
```

## ITEM FEATURE MATRIX WITH RATING

```python
for i in vect.get_feature_names():
    df_item_features[i]= df_item_features['rating'] * df_item_features[i]

df_item_features
```

## CALCULATING USER FEATURE VECTOR

```python
df_item_features[vect.get_feature_names()].sum()
```

```python
user_feature_vector = df_item_features[vect.get_feature_names()].sum()/df_item_features[vect.get_feature_names()].sum().sum()
user_feature_vector
```

## RECOMMENDATION

```python
recommend = df_genre.loc[~df_genre['name'].isin([
    'Kimi no Na wa.', 
    'Fullmetal Alchemist: Brotherhood', 
    'Gintama°'
    ])] 
```

```python
for i in vect.get_feature_names():
    recommend[i]= recommend[i] * user_feature_vector[i]
recommend
```

## PREDICTED SCORE COLUMN

```python
recommend['predicted_score'] = recommend.drop(columns=['rating', 'anime_id']).sum(axis=1)
```

```python
recommend
```

```python
recommend[['name', 'predicted_score']].sort_values('predicted_score', ascending=False).head(20)
```
