## **Data Preprocessing**

### **Drop CUST_ID**

Karena Customer ID adalah ID unik untuk tiap customer, maka fitur ini tidak memiliki peranan penting untuk penentuan cluster nantinya.

```python
df.drop(['CUST_ID'], axis=1, inplace=True)
```

### **Drop missing value pada CREDIT_LIMIT**



```python
df.dropna(subset=['CREDIT_LIMIT'], inplace=True)
```

### **Imputasi missing value pada MINIMUM_PAYMENTS**

Karena distribusi pada MINIMUM_PAYMENTS tidak mengikuti distribusi normal (terdapat skewness), maka mengimputasi missing value dengan menggunakan nilai median akan memberikan estimasi nilai yang lebih baik.

```python
df['MINIMUM_PAYMENTS'].fillna(df['MINIMUM_PAYMENTS'].median(), inplace=True)
```

### **Scaling data**

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
```

```python
df_scaled = pd.DataFrame(df_scaled)
df_scaled.columns = df.columns
df_scaled
```

```python
df_scaled.describe()
```

```python
plt.figure(figsize=(20,35))
for i, col in enumerate(df_scaled.columns):
    if df_scaled[col].dtype != 'object':
        ax = plt.subplot(9, 2, i+1)
        sns.kdeplot(df_scaled[col], ax=ax)
        plt.xlabel(col)
        
plt.show()
```

### **Interpretasi**

Meskipun masih terdapat skewness pada data, namun dengan melakukan standarisasi, kita sudah meminimalkan skala dari tiap feature sehingga bisa memiliki nilai mean = 0 dan std = 1.

## **Dimensionality reduction (PCA)**

```python
pca = PCA(random_state=0)
pca.fit(df_scaled)
```

```python
sns.set_style('whitegrid')
plt.plot(pca.explained_variance_ratio_.cumsum());
```

```python
# Variance explained with the number of features
pca = PCA(random_state=0)
pca.fit(df_scaled)
features = range(pca.n_components_)

plt.figure(figsize=(8,4))
plt.bar(features[:16], pca.explained_variance_ratio_[:16], color='lightskyblue')
plt.xlabel('PCA feature')
plt.ylabel('Variance')
plt.xticks(features[:16]);
```

```python
pca.explained_variance_ratio_[0:2].sum()
```

```python
pca = PCA(n_components=2, random_state=0)
X = pca.fit_transform(df_scaled)
```

## **K-Means clustering**

### **Elbow method**

```python
kmeans_models = [KMeans(n_clusters=k, random_state=0).fit(X) for k in range (1, 10)]
inertia = [model.inertia_ for model in kmeans_models]

plt.plot(range(1, 10), inertia)
plt.title('Elbow method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS');
```

```python
silhoutte_scores = [silhouette_score(X, model.labels_) for model in kmeans_models[2:6]]
plt.plot(range(2, 6), silhoutte_scores, 'bo-')
plt.xticks([2, 3, 4, 5])
plt.title('Silhoutte scores vs Number of clusters')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette score');
```

### **Silhouette method**

```python
kmeans = KMeans(random_state=0)
kmeans.fit(X)
```

```python
labels = kmeans.labels_
silhouette_score(X, labels, metric = 'euclidean', random_state=0)
```

```python
silhouette_scores = []
number_of_cluster = range(2, 11)

for i in number_of_cluster:
  kmeans = KMeans(n_clusters=i, random_state=0)
  kmeans.fit(X)

  labels = kmeans.labels_
  silhouette_scores.append(silhouette_score(X, labels, metric = 'euclidean', random_state=0))
```

```python
plt.figure(figsize=(12,7))

sns.lineplot(number_of_cluster, silhouette_scores)
sns.scatterplot(number_of_cluster, silhouette_scores)

plt.xticks(number_of_cluster)

plt.xlabel('Number of clusters')
plt.ylabel('Silhouette score');
```

```python
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X)
labels = kmeans.labels_
```

```python
pca_df = pd.DataFrame(data=X, columns=['PC1', 'PC2'])
pca_df['labels'] = labels
pca_df.head()
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x='PC1', y='PC2', hue='labels', data=pca_df, palette='bright');
```

### **Interpretasi**

Jika kita menggunakan PCA pada clustering, tentu hasil model menjadi lebih sulit diinterpretasi karena principal component sederhananya adalah kombinasi dari berbagai fitur. Untuk dataset Creditcard dan hasil processing yang telah dilakukan, didapat bahwa jumlah cluster terbaik untuk n_components=2 adalah 2 cluster. 

Masih banyak yang dapat diimprovisasi agar bisa mendapatkan nilai variance yang lebih besar untuk 2 principal component pertama, karena hasil yang didapat saat ini hanya sekitar 47% variance saja yang terwakili. Hal ini dapat dipengaruhi oleh karena adanya outliers pada data.
